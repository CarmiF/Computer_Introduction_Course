```python
Arr1 = [21,80,99,49,18,81,77,13]
Arr2 = [31,13,88,65,25,10,52,40]
SIZE = len(Arr1)
mod4_ID1 = []
mod4_ID2 = []

for i in range(SIZE):
  if Arr1[i]%4:
    mod4_ID1.append(-1)
  else:
    mod4_ID1.append(0)

  if Arr2[i]%4:
    mod4_ID2.append(-1)
  else:
    mod4_ID2.append(0)


print(mod4_ID1)
print(mod4_ID2)
```
